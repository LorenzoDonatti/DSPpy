```python
M = [10,25,50,101]
w = np.linspace(-np.pi, np.pi, 201)

X1,X2,X3,X4 = [],[],[],[]

for m in M:
    n = np.arange(0,m)

    x1 = np.ones(m)
    X1.append(dtft(x1,n,w))

    x2 = 0.5 * (1 - np.cos(2*np.pi*n / (m-1))) * x1
    X2.append(dtft(x2,n,w))

    x3 = (1 - np.abs(m - 1 - 2*n)/(m-1)) * x1
    X3.append(dtft(x3,n,w))

    x4 = (0.54 - 0.46 * np.cos(2*np.pi*n / (m-1))) * x1
    X4.append(dtft(x4,n,w))

# Criar uma nova figura com subplots
fig, axs = plt.subplots(4, 4, figsize=(14, 10))

axs[0, 0].plot(w/np.pi, abs(X1[0])/max(abs(X1[0]))), axs[0, 1].plot(w/np.pi, abs(X1[1])/max(abs(X1[1]))), axs[0, 2].plot(w/np.pi, abs(X1[2])/max(abs(X1[2]))), axs[0, 3].plot(w/np.pi, abs(X1[3])/max(abs(X1[3])))
axs[0, 0].set_title('Rectangular - M = 10'), axs[0, 1].set_title('Rectangular - M = 25'), axs[0, 2].set_title('Rectangular - M = 50'), axs[0, 3].set_title('Rectangular - M = 101')

axs[1, 0].plot(w/np.pi, abs(X2[0])/max(abs(X2[0]))), axs[1, 1].plot(w/np.pi, abs(X2[1])/max(abs(X2[1]))), axs[1, 2].plot(w/np.pi, abs(X2[2])/max(abs(X2[2]))), axs[1, 3].plot(w/np.pi, abs(X2[3])/max(abs(X2[3])))
axs[1, 0].set_title('Hanning - M = 10'), axs[1, 1].set_title('Hanning - M = 25'), axs[1, 2].set_title('Hanning - M = 50'), axs[1, 3].set_title('Hanning - M = 101')

axs[2, 0].plot(w/np.pi, abs(X3[0])/max(abs(X3[0]))), axs[2, 1].plot(w/np.pi, abs(X3[1])/max(abs(X3[1]))), axs[2, 2].plot(w/np.pi, abs(X3[2])/max(abs(X3[2]))), axs[2, 3].plot(w/np.pi, abs(X3[3])/max(abs(X3[3])))
axs[2, 0].set_title('Triangular - M = 10'), axs[2, 1].set_title('Triangular - M = 25'), axs[2, 2].set_title('Triangular - M = 50'), axs[2, 3].set_title('Triangular - M = 101')

axs[3, 0].plot(w/np.pi, abs(X4[0])/max(abs(X4[0]))), axs[3, 1].plot(w/np.pi, abs(X4[1])/max(abs(X4[1]))), axs[3, 2].plot(w/np.pi, abs(X4[2])/max(abs(X4[2]))), axs[3, 3].plot(w/np.pi, abs(X4[3])/max(abs(X4[3])))
axs[3, 0].set_title('Hamming - M = 10'), axs[3, 1].set_title('Hamming - M = 25'), axs[3, 2].set_title('Hamming - M = 50'), axs[3, 3].set_title('Hamming - M = 101')

plt.tight_layout()
plt.show()
```

```python
Dt = 0.001
t = np.arange(0,1,Dt)

ns = [np.arange(0,101), np.arange(0,21), np.arange(0,11)]

Xa = []
for T,n in zip(Ts, ns):
    n = n.reshape(-1,1)
    x1 = np.cos(20*np.pi*n*T)
    Fs = (1/T) * np.ones([len(n),1])
    nT = (n * T) * np.ones([1,len(t)])
    xa = (x1.T @ np.sinc(Fs * (t - nT))).T
    Xa.append(xa)

# Criar uma nova figura com subplots
fig, axs = plt.subplots(3, 1, figsize=(6, 8))
# Plotar o sinal original na primeira linha
axs[0].plot(t, Xa[0])
axs[0].set_title('Sinc Interpolation:    Ts = 0.01s')
axs[0].set_xlabel('n')
axs[0].set_ylabel('x(n)')
axs[0].set_ylim(-1.2, 1.2)

# Plotar a parte par na segunda linha
axs[1].plot(t, Xa[1])
axs[1].set_title('Sinc Interpolation:    Ts = 0.05s')
axs[1].set_xlabel('n')
axs[1].set_ylabel('x(n)')
axs[1].set_ylim(-1.2, 1.2)

# Plotar a parte par na segunda linha
axs[2].plot(t, Xa[2])
axs[2].set_title('Sinc Interpolation:    Ts = 0.1s')
axs[2].set_xlabel('n')
axs[2].set_ylabel('x(n)')
axs[2].set_ylim(-1.2, 1.2)

plt.tight_layout()
plt.show()
```

```python
from scipy.interpolate import CubicSpline
Dt = 0.001
t = np.arange(0,1,Dt)

ns = [np.arange(0,101), np.arange(0,21), np.arange(0,11)]

Xa = []
for T,n in zip(Ts, ns):
    x1 = np.cos(20*np.pi*n*T)
    xa = CubicSpline(x=n*T, y=x1)
    Xa.append(xa(t))


# Criar uma nova figura com subplots
fig, axs = plt.subplots(3, 1, figsize=(6, 8))
# Plotar o sinal original na primeira linha
axs[0].plot(t, Xa[0])
axs[0].set_title('Sinc Interpolation:    Ts = 0.01s')
axs[0].set_xlabel('n')
axs[0].set_ylabel('x(n)')
axs[0].set_ylim(-1.2, 1.2)

# Plotar a parte par na segunda linha
axs[1].plot(t, Xa[1])
axs[1].set_title('Sinc Interpolation:    Ts = 0.05s')
axs[1].set_xlabel('n')
axs[1].set_ylabel('x(n)')
axs[1].set_ylim(-1.2, 1.2)

# Plotar a parte par na segunda linha
axs[2].plot(t, Xa[2])
axs[2].set_title('Sinc Interpolation:    Ts = 0.1s')
axs[2].set_xlabel('n')
axs[2].set_ylabel('x(n)')
axs[2].set_ylim(-1.2, 1.2)

plt.tight_layout()
plt.show()
```
