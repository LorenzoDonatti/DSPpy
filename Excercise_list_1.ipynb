P2.5

Um sinal periódico segue esta suposição:
> x(n + p) = x(n)

Considere:

> x(n) = cos(2pi f n)

Para que este seja periódico, devemos ser capazes de encontrar um inteiro não nulo p tal que para todos os inteiros n,

> cos(2pi f (n + p)) = cos(2pi f n)

Isso só pode ser verdadeiro se 2pi f p for um múltiplo de 2pi. Ou seja, se houver algum inteiro m tal que:

> 2pi f p = 2pi m

Dividindo ambos os lados por 2pi p, vemos que este sinal é periódico apenas se pudermos encontrar inteiros não nulos p e m tais que:

> f = m/p

Em outras palavras, f deve ser racional. Somente se f for racional este sinal será periódico.


```python
n = np.arange(-100,100)
x1 = np.exp(0.1j * np.pi * n)

# Criar uma nova figura com subplots
fig, axs = plt.subplots(1, 1, figsize=(10, 4))

# Plotar o sinal original na primeira linha
axs.stem(n, x1.imag)
axs.set_title('Sinal Original')
axs.set_xlabel('Índice')
axs.set_ylabel('Valor')

print('Considerando f0 = w0 / 2pi == K/N e que w0 = 0.1pi, f0 = 0.1/2 == 1/20.')
print('Ou seja, K = 1 e N = 20, sendo a função Periódica.')
```

```python
n = np.arange(-20,20)
w0 = 0.1
x1 = np.cos(w0 * n)

f0 = w0/(2 * np.pi)

# Criar uma nova figura com subplots
fig, axs = plt.subplots(1, 1, figsize=(10, 4))

# Plotar o sinal original na primeira linha
axs.stem(n, x1)
axs.set_title('Sinal Original')
axs.set_xlabel('Índice')
axs.set_ylabel('Valor')


print(f"A função não é periodica, visto que o valor de f0 é {w0}/2pi == {f0}, sendo não racional.")
```

```python
#Essa função consegue reproduzir os valores do np.correlate(mode='valid'), mas não o np.correlate(mode='full')
def ccorr(x, y):
    corr = np.zeros(len(x) - len(y) + 1)  # Inicializa o array de autocorrelação com zeros
    corr[0] = np.sum(x[:len(y)] * y)
    print(corr[0])
    for l in range(1,len(corr)):
        corr[l] = np.sum(x[l:len(y) + l] * y)  # Para l > 0, move a janela ao longo de x e calcula a autocorrelação
    return corr
```

```python
nx = np.arange(21)
ny = np.arange(-20, 1)

x = 0.9 ** nx 
y = 0.8 ** (-ny)

#corr = ccorr(x,y) readaptar

acorr_x = conv_m_to_corr(x,nx,x,nx)
acorr_y = conv_m_to_corr(y,ny,y,ny)
corr_xy = conv_m_to_corr(x,nx,y,ny)

# Criar uma nova figura com subplots
fig, axs = plt.subplots(5, 1, figsize=(6, 8))
# Plotar o sinal original na primeira linha
axs[0].stem(nx, x)
axs[0].set_title('Sinal x')
axs[0].set_xlabel('Índice')
axs[0].set_ylabel('Valor')

# Plotar a parte par na segunda linha
axs[1].stem(ny, y)
axs[1].set_title('Sinal y')
axs[1].set_xlabel('Índice')
axs[1].set_ylabel('Valor')

# Plotar o sinal original na primeira linha
axs[2].stem(acorr_x[1], acorr_x[0])
axs[2].set_title('Autocorrelação x')
axs[2].set_xlabel('Índice')
axs[2].set_ylabel('Valor')

# Plotar a parte par na segunda linha
axs[3].stem(acorr_y[1], acorr_y[0])
axs[3].set_title('Autocorrelação y')
axs[3].set_xlabel('Índice')
axs[3].set_ylabel('Valor')

# Plotar a parte par na segunda linha
axs[4].stem(corr_xy[1], corr_xy[0])
axs[4].set_title('Correlação x e y')
axs[4].set_xlabel('Índice')
axs[4].set_ylabel('Valor')

plt.tight_layout()
plt.show()
```

```python
n1 = np.arange(-5,25)
x1 = np.cos(np.pi*n1/4)*(stepseq(-5,n1[0],n1[-1])[0] - stepseq(25,n1[0],n1[-1])[0])

n2 = np.arange(-0,20)
x2 = (0.9**(-n2)) * (stepseq(0, n2[0],n2[-1])[0] - stepseq(20, n2[0],n2[-1])[0])

n3 = np.arange(-10,10)
x3 = np.round(5 * np.random.uniform(n3/10))

#COMUTAÇÃO
conv12,n12 = conv_m(x1,n1,x2,n2)
conv21,n21 = conv_m(x2,n2,x1,n1)

conv13,n13 = conv_m(x1,n1,x3,n3)
conv31,n31 = conv_m(x3,n3,x1,n1)

conv23,n23 = conv_m(x2,n2,x3,n3)
conv32,n32 = conv_m(x3,n3,x2,n2)

#ASSOCIAÇÃO
conv123, n123 = conv_m(conv12,n12,x3,n3)
conv321, n321 = conv_m(conv32,n32,x1,n1)

# DISTRIBUIÇÃO
x23,nd23 = sigadd(x2,n2,x3,n3)

dist1,nd1 = conv_m(x1,n1,x23,nd23)
dist2,nd2 = sigadd(conv12,n12,conv13,n13)

#IDENTIDADE
ndelta = np.arange(-1,5)
delta = impseq(-1, ndelta[0], ndelta[-1])

ident1,ni1 = conv_m(x1,n1,delta[0],ndelta)
x1_shift,ns1 = sigshift(x1,n1, -1)

ident2,ni2 = conv_m(x2,n2,delta[0],ndelta)
x2_shift,ns2 = sigshift(x2,n2, -1)

ident3,ni3 = conv_m(x3,n3,delta[0],ndelta)
x3_shift,ns3 = sigshift(x3,n3, -1)


# Criar uma nova figura com subplots
fig, axs = plt.subplots(8, 2, figsize=(15, 10))

# Plotar o sinal original na primeira linha
axs[0, 0].stem(n12, conv12), axs[0, 1].stem(n21, conv21)
axs[0, 0].set_title('Convolução x1 e x2'), axs[0, 1].set_title('Convolução x2 e x1')

# Plotar o sinal original na primeira linha
axs[1, 0].stem(n13, conv13), axs[1, 1].stem(n31, conv31)
axs[1, 0].set_title('Convolução x1 e x3'), axs[1, 1].set_title('Convolução x3 e x1')

# Plotar o sinal original na primeira linha
axs[2, 0].stem(n23, conv23), axs[2, 1].stem(n32, conv32)
axs[2, 0].set_title('Convolução x2 e x3'), axs[2, 1].set_title('Convolução x3 e x2')

#ASSOCIAÇÃO GRAFICO

# Plotar o sinal original na primeira linha
axs[3, 0].stem(n123, conv123), axs[3, 1].stem(n321, conv321)
axs[3, 0].set_title('Convolução [x1 e x2] e x3 '), axs[3, 1].set_title('Convolução [x2 e x3] e x1')


#DISTRIBUIÇÃO GRAFICO

# Plotar a parte par na segunda linha
axs[4, 0].stem(nd1, dist1), axs[4, 1].stem(nd2, dist2)
axs[4, 0].set_title('Convolução x1 e [x2 + x3]'), axs[4, 1].set_title('Convolução x1 e x2 + Convolução x1 e x3')


#IDENTIDADE GRAFICO 
# Plotar o sinal original na primeira linha
axs[5, 0].stem(ni1, ident1), axs[5, 1].stem(ns1, x1_shift)
axs[5, 0].set_title('Convolução x1 e delta(n-n0)'), axs[5, 1].set_title('x1(n-n0)')

# Plotar a parte par na segunda linha
axs[6, 0].stem(ni2, ident2), axs[6, 1].stem(ns2, x2_shift)
axs[6, 0].set_title('Convolução x2 e delta(n-n0)'), axs[6, 1].set_title('x2(n-n0)')

# Plotar a parte par na segunda linha
axs[7, 0].stem(ni3, ident3), axs[7, 1].stem(ns3, x3_shift)
axs[7, 0].set_title('Convolução x3 e delta(n-n0)'), axs[7, 1].set_title('x3(n-n0)')

plt.tight_layout()
plt.show()
```

```python
def simple_differentiatior(x):
    y = np.zeros(len(x))
    y[0] = x[0]
    y[1:] = x[1:] - x[0:-1]

    return y
```

```python
#Diferenciator formula
#y(n) = x(n) - x(n-1)

n = np.arange(-50,50)

x1 = 5*(stepseq(0,n[0],n[-1])[0] - stepseq(20,n[0],n[-1])[0])
y1 = simple_differentiatior(x1)

# Criar uma nova figura com subplots
fig, axs = plt.subplots(2, 1, figsize=(5, 6))

# Plotar o sinal original na primeira linha
axs[0].stem(n, x1)
axs[0].set_title('Sinal Original')
axs[0].set_xlabel('Índice')
axs[0].set_ylabel('Valor')

# Plotar a parte par na segunda linha
axs[1].stem(n, y1)
axs[1].set_title('Sinal Diferenciado')
axs[1].set_xlabel('Índice')
axs[1].set_ylabel('Valor')


plt.tight_layout()
plt.show()
```
